```c++
#include<iostream>
#include<string>
#include<stdio.h>
```

```c++
TString muon = "muon.root";
TString bottom = "bottom.root";
TString top = "top.root";
TString charm = "charm.root";
TString gluon = "gluon.root";
TString foton = "foton.root";
TString W = "W.root";
TString Z = "Z.root";
```

```c++
TChain chain("Tree;1");
chain.Add(muon);
chain.Add(bottom);
chain.Add(top);
chain.Add(charm);
chain.Add(gluon);
chain.Add(foton);
chain.Add(W);
chain.Add(Z);
```

```c++
Float_t m;
Float_t E;
Float_t pT;
chain.SetBranchAddress("m", &m);
chain.SetBranchAddress("E", &E);
chain.SetBranchAddress("pT", &pT);
```

```c++
TCanvas *canvas = new TCanvas("Canvas","a first way to plot a variable",800,600);
```

```c++
TH1F *hist_pt = new TH1F("pT","pT: Transverse Momentum; pT ; Events ",50,0,1000);
TH1F *hist_m = new TH1F("m","m: mass; m ; Events ",50,0,200);
TH1F *hist_E = new TH1F("E","E: Energy; E ; Events ",50,0,1000);
```

```c++
int nentries, nbytes, i;
nentries = (Int_t)chain.GetEntries();

for (i = 0; i < nentries; i++)
{
    nbytes = chain.GetEntry(i);
    // Variable pT
    hist_pt->Fill(pT);
    // Variable m
    hist_m->Fill(m);
    // Variable E
    hist_E->Fill(E);
}

std::cout << "Done!" << std::endl;
```

```c++
hist_pt->SetFillColor(kRed);
hist_pt->Draw();
canvas->Draw();
```

```c++
hist_E->SetFillColor(kRed);
hist_E->Draw();
canvas->Draw();
```
